\begin{equation} D_{x}\ and\ D_{y}\ is\ (row*column)\ *\ (row*column)\ matrix \end{equation}
\begin{equation}D_{x}=\begin{Bmatrix} 
     -1 & 0 & \dots & 1 & 0 &\dots & 0 \\
    0 & -1 & \dots & 0&1 & \dots & 0 \\
    \vdots & \vdots & \vdots & \vdots & \\
   \end{Bmatrix}
   \end{equation}

\begin{equation}D_{y}=\begin{Bmatrix} 
     -1 & 1 & \dots & 0&0 \\
     0 &  -1 & 1& \dots & 0 \\
    \vdots & \vdots & \vdots & \vdots & \\
   \end{Bmatrix}
   \end{equation}
- use least square, find u
\begin{equation}\|\begin{Bmatrix} 
     I \\
     D_{x} \\
    D_{y} \\
   \end{Bmatrix}u-\begin{Bmatrix} 
     f \\
     0 \\
     0 \\
   \end{Bmatrix}\|
   \end{equation}

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from skimage import io, color
from skimage import exposure
import numpy.linalg as lin

file_image	= 'cau.jpg'

im_color 	= io.imread(file_image)
im_gray  	= color.rgb2gray(im_color)
im          = (im_gray - np.mean(im_gray)) / np.std(im_gray)

(row, col)  = im.shape

def x_deravative_matrix(row, column):
    n = row * column
    result = np.zeros((n, n))
    for i in range(n):
        if i<n-column:
            result[i][i] = -1
            result[i][i+column] = 1
        else:
            result[i][i] = 1            
    return result

def y_deravative_matrix(row, column):
    n = row * column
    result = np.zeros((n, n))
    for i in range(n):
        if i!=n-1:
            result[i][i+1] = 1
            result[i][i] = -1
        else:
            result[i][i] = 1
    return result



for i in range(1,5):
    for j in range(-10,10):
        
        noise_std   = 0.2 * i # try with varying noise standard deviation
        noise       = np.random.normal(0, noise_std, (row, col))
        im_noise    = im + noise
        u_vector = im_noise.reshape(-1)
        identity_matrix = np.identity(row*col)
            # obtain the reconstructed (denoised) images with varying degrees of regularization 
        _lambda = 2**j
        x_der = _lambda * x_deravative_matrix(row, col)
        y_der = _lambda * y_deravative_matrix(row, col)
        identity_matrix = np.identity(row*col)
        f = im_noise.reshape(-1)
        zero_matrix = np.zeros((row*col))
        A = np.concatenate((identity_matrix ,x_der, y_der), axis = 0)
        B = np.concatenate((f, zero_matrix, zero_matrix), axis = 0)
        u = np.dot(np.dot(lin.inv(np.dot(A.T, A)), A.T), B)
        u = u.reshape(row,col)
        
        im_recon    = u # this must be replaced with the reconstructed (denoised) image 
        noise_recon = im_noise - im_recon
        error       = np.linalg.norm(im - im_recon)

        print('std:',noise_std,',lambda',_lambda,',error', error)
        p1 = plt.subplot(2,2,1)
        p1.set_title('original image')
        plt.imshow(im_gray, cmap='gray')
        plt.axis('off')
      
        
        p2 = plt.subplot(2,2,2)
        p2.set_title('noisy image')
        plt.imshow(im_noise, cmap='gray')
        plt.axis('off')
       
        
        p3 = plt.subplot(2,2,3)
        p3.set_title('reconstruction')
        plt.imshow(u, cmap='gray')
        plt.axis('off')
      
        p4 = plt.subplot(2,2,4)
        p4.set_title('estimated noise')
        plt.imshow(noise_recon, cmap='gray')
        plt.axis('off')
        plt.show()
```
